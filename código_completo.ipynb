```python
#login
import numpy as np

v_usuarios = np.array(["ana", "bruno", "carla"])
v_senhas = np.array(["1234", "abcd", "senha"])
usuario_logado = None
cont = 0

while True:
    if usuario_logado:
        print(f"\nUsuário logado: {usuario_logado.capitalize()}")
        break

    else:
        if cont >= 3:
            print("Número máximo de tentativas excedido. Saindo...")
            break

        print("\nLogin")
        print("1. Entrar")
        print("2. Criar novo usuário")
        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            usuario_input = input('Digite seu usuário: ').lower()
            senha_input = input('Digite sua senha: ')

            if usuario_input in v_usuarios:
                idx = np.where(v_usuarios == usuario_input)[0][0]
                if v_senhas[idx] == senha_input:
                    print(f"Acesso permitido. Seja bem-vindo(a), {usuario_input.capitalize()}!")
                    usuario_logado = usuario_input
                    cont = 0
                else:
                    print("Senha incorreta.")
                    cont += 1
            else:
                print("Usuário não encontrado.")
                cont += 1

        elif escolha == '2':
            novo_usuario = input("Digite o novo nome de usuário: ").lower()
            if novo_usuario in v_usuarios:
                print("Esse nome de usuário já existe. Tente outro.")
            else:
                nova_senha = input("Digite a nova senha: ")
                v_usuarios = np.append(v_usuarios, novo_usuario)
                v_senhas = np.append(v_senhas, nova_senha)
                usuario_logado = novo_usuario
                cont = 0
                print(f"Usuário {novo_usuario.capitalize()} criado com sucesso e logado!")
            break

        else:
            print("Opção inválida!")

#Dados de usuários logados
usuarios = [
  ["ana", "1234", True],
  ["bruno", "abcd", True],
  ["carla", "senha", True]
]

while True:
  nome = input("Digite o nome do usuário (ou 'sair' para encerrar): ")
  if nome.lower() == "sair":
    break

  senha = input("Digite a senha: ")
  logado_status = False
  for u in usuarios:
    if u[0].lower() == nome.lower():
      logado_status = True
      if u[1] == senha:
        if u[2]:
          print(f"{u[0]} está logado.")

        else:
          print(f"{u[0]} não está logado.")
      else:
        print("Senha incorreta.")
      break

  if not logado_status:
    print("Usuário não encontrado.")

#lista de produtos
produtos = [
    {"nome": "Perfume Floral", "categoria": "Perfumes", "preco": 129.90},
    {"nome": "Colônia Masculina", "categoria": "Perfumes", "preco": 99.50},
    {"nome": "Creme Hidratante", "categoria": "Cuidados com a Pele", "preco": 45.00},
    {"nome": "Sabonete Líquido", "categoria": "Higiene", "preco": 22.30},
    {"nome": "Loção Pós-Barba", "categoria": "Barbear", "preco": 38.90},
    {"nome": "Shampoo Reconstrutor", "categoria": "Cabelos", "preco": 29.90},
    {"nome": "Condicionador", "categoria": "Cabelos", "preco": 28.50},
    {"nome": "Óleo Corporal", "categoria": "Cuidados com a Pele", "preco": 58.00},
    {"nome": "Desodorante Roll-on", "categoria": "Higiene", "preco": 14.20},
    {"nome": "Máscara Capilar", "categoria": "Cabelos", "preco": 67.80}
]
print(produtos)

def mostrar_categorias():
    categorias = set()
    for produto in produtos:
        categorias.add(produto["categoria"])
    print("Categorias disponíveis:")
    for categoria in categorias:
        print(f"- {categoria}")

def mostrar_produtos_da_categoria(categoria_escolhida):
    encontrou = False
    print(f"\nProdutos da categoria '{categoria_escolhida}':")
    for produto in produtos:
        if produto["categoria"].lower() == categoria_escolhida.lower():
            print(f"- {produto['nome']} | Preço: R$ {produto['preco']:.2f}")
            encontrou = True
    return encontrou

def main():
    while True:
        mostrar_categorias()
        categoria = input("\nDigite o nome da categoria desejada (ou 'sair' para encerrar): ")
        if categoria.lower() == 'sair':
            break
        if not mostrar_produtos_da_categoria(categoria):
            print("Categoria inválida ou sem produtos. Tente novamente.")

main()

#Carrinho de compras
from collections import deque

fila = []
cont = 0

while cont < 15:
  n_produto = input("Digite o produto que deseja adicionar ao carrinho ou se deseja sair: ")
  fila.append(n_produto)
  if n_produto != 'sair':
   print(f'Produto', n_produto, 'adicionado ao carrinho.')
  cont += 1
  if n_produto == 'sair':
    break
    print(fila)
print("1 - Realizar a compra")
print("2 - Continuar comprando")
compra = input("Digite a opção desejada: ")
if compra == '1':
    fila.clear()
    print("Compra realizada com sucesso!")

else:
 print("Continuar comprando")
 while cont < 15:
  n_produto = input("Digite o produto que deseja adicionar ao carrinho ou se deseja sair: ")
  fila.append(n_produto)
  if n_produto != 'sair':
   print(f'Produto', n_produto, 'adicionado ao carrinho.')
  cont += 1
  if n_produto == 'sair':
    break
    print(fila)

print(fila)
#forma de pagamento

formas_pagamento = ["PIX", "Cartão", "Boleto"]
historico_pagamentos = []

while True:
    print("\nFORMAS DE PAGAMENTO")
    print("1. Efetuar pagamento")
    print("2. Sair")
    print("3. Ver último pagamento tentado")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        print("Escolha a forma de pagamento:")
        print("1. PIX")
        print("2. Cartão")
        print("3. Boleto")
        escolha = input("Digite o número referente à forma de pagamento desejada: ")

        if escolha == "1":
            nome = input("Digite seu nome: ")
            chave = input("Digite a chave PIX: ")
            valor = input("Digite o valor: ")
            print(f"\nVocê vai pagar R${valor} via PIX para {nome}, chave: {chave}")
            confirma = input("Confirma o pagamento? (s/n): ")

            pagamento = {
                "tipo": "PIX",
                "nome": nome,
                "valor": valor,
                "status": "Confirmado" if confirma.lower() == "s" else "Cancelado"
            }
            historico_pagamentos.append(pagamento)

            if confirma.lower() == "s":
                print("Pagamento via PIX feito com sucesso!")
                break
            else:
                print("Pagamento cancelado.")

        elif escolha == "2":
            nome = input("Digite seu nome: ")
            cartao = input("Número do cartão: ")
            tipo = input("Digite '1' para crédito ou '2' para débito: ")
            valor = input("Digite o valor: ")

            if tipo == "1":
                parcelas = input("Em quantas parcelas deseja pagar? ")
                print(f"\nVocê vai pagar R${valor} em {parcelas}x no cartão de crédito para {nome}, cartão: ****{cartao[-4:]}")
            elif tipo == "2":
                parcelas = "1"
                print(f"\nVocê vai pagar R${valor} no cartão de débito para {nome}, cartão: ****{cartao[-4:]}")
            else:
                print("Tipo de cartão inválido.")
                continue

            confirma = input("Confirma o pagamento? (s/n): ")

            pagamento = {
                "tipo": "Cartão (Crédito)" if tipo == "1" else "Cartão (Débito)",
                "nome": nome,
                "valor": valor,
                "parcelas": parcelas,
                "status": "Confirmado" if confirma.lower() == "s" else "Cancelado"
            }
            historico_pagamentos.append(pagamento)

            if confirma.lower() == "s":
                print("Pagamento com cartão feito com sucesso!")
                break
            else:
                print("Pagamento cancelado.")

        elif escolha == "3":
            nome = input("Digite seu nome: ")
            cpf = input("Digite seu CPF: ")
            valor = input("Digite o valor: ")
            parcelas = input("Em quantas parcelas deseja pagar o boleto? ")
            print(f"\nVocê vai gerar boleto no valor de R${valor} dividido em {parcelas}x para {nome}, CPF: {cpf}")
            confirma = input("Confirma o pagamento? (s/n): ")

            pagamento = {
                "tipo": "Boleto",
                "nome": nome,
                "valor": valor,
                "parcelas": parcelas,
                "status": "Confirmado" if confirma.lower() == "s" else "Cancelado"
            }
            historico_pagamentos.append(pagamento)

            if confirma.lower() == "s":
                print("Boleto gerado com sucesso!")
                break
            else:
                print("Pagamento cancelado.")

        else:
            print("Opção inválida.")

    elif opcao == "2":
        print("Saindo...")
        break

    elif opcao == "3":
        if historico_pagamentos:
            ultimo = historico_pagamentos[-1]
            print("\nÚltimo pagamento registrado:")
            for chave, valor in ultimo.items():
                print(f"{chave.capitalize()}: {valor}")
        else:
            print("Nenhum pagamento registrado ainda.")
    else:
        print("Opção inválida.")

#logout

while True:
    if usuario_logado:
        opcao = input(f"\nVocê está logado como {usuario_logado.capitalize()}. Deseja fazer logout? (s/n): ").lower()
        if opcao == 's':
            print(f"Usuário {usuario_logado.capitalize()} deslogado com sucesso!")
            usuario_logado = None
        elif opcao == 'n':
            print("Continuando logado.")
            break
        else:
            print("Resposta inválida. Digite 's' ou 'n'.")
    else:
        print("Nenhum usuário está logado.")
        break

```
